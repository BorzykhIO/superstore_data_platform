```python
# Инициализация Faker для генерации случайных данных
import pandas as pd
import random
import numpy as np
from faker import Faker

faker = Faker()

# Связь категорий и подкатегорий
category_subcategory_mapping = {
    "Furniture": ["Bookcases", "Chairs", "Tables", "Furnishings"],
    "Office Supplies": ["Labels", "Storage", "Art", "Binders", "Appliances", "Paper", "Envelopes", "Fasteners", "Supplies"],
    "Technology": ["Phones", "Accessories", "Machines", "Copiers"]
}

# Количество строк для генерации
num_rows = 1000

# Генерация данных
rows = []

for _ in range(num_rows):
    order_id = f"{faker.random_element(['CA', 'US', 'EU'])}-{random.randint(2014, 2017)}-{random.randint(100000, 999999)}"
    row_id = random.randint(1, 9999)
    order_date = faker.date_between(start_date="-3y", end_date="today")
    ship_date = order_date + pd.Timedelta(days=random.randint(1, 7))
    ship_mode = random.choice(["Standard Class", "Second Class", "First Class", "Same Day"])

    customer_id = f"{faker.random_uppercase_letter()}{faker.random_uppercase_letter()}-{random.randint(10000, 99999)}"
    customer_name = faker.name()
    segment = random.choice(["Consumer", "Corporate", "Home Office"])

    country = "United States"
    city = faker.city()
    state = faker.state()
    postal_code = faker.zipcode()
    region = random.choice(["West", "East", "Central", "South"])

    category = random.choice(list(category_subcategory_mapping.keys()))
    sub_category = random.choice(category_subcategory_mapping[category])

    product_id = f"{random.choice(['FUR', 'OFF', 'TEC'])}-{faker.random_uppercase_letter()}{faker.random_uppercase_letter()}-{random.randint(10000000, 99999999)}"
    product_name = faker.catch_phrase()

    sales = round(random.uniform(5.0, 500.0), 2)
    quantity = random.randint(1, 10)
    discount = round(random.uniform(0.0, 0.5), 2)
    profit = round(sales * (1 - discount) - random.uniform(5.0, 50.0), 2)

    rows.append([
        row_id, order_id, order_date, ship_date, ship_mode, customer_id, customer_name, segment,
        country, city, state, postal_code, region, product_id, category, sub_category, product_name,
        sales, quantity, discount, profit
    ])

# Создание DataFrame
columns = [
    "Row ID", "Order ID", "Order Date", "Ship Date", "Ship Mode", "Customer ID", "Customer Name", "Segment",
    "Country", "City", "State", "Postal Code", "Region", "Product ID", "Category", "Sub-Category", "Product Name",
    "Sales", "Quantity", "Discount", "Profit"
]

synthetic_data = pd.DataFrame(rows, columns=columns)

# Сохранение данных в CSV
# synthetic_data.to_csv("synthetic_superstore.csv", index=False, encoding="utf-8")

print("Синтетические данные успешно сгенерированы и сохранены в 'synthetic_superstore.csv'.")
```
